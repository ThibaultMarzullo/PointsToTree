```python
import pprint
import re

from pointstotree.point_tree import PointTree
```

```python
"""Here are our points"""

file_path = "../dataset/Gabe_Fierro-An_Open_Dataset_and_Collection_Tool_for_BMS_Point_Labels-Sample.csv"
with open(file_path, "r") as f:
    points = f.read().split("\n")

pprint.pprint(points)
```

```python
"""For this specific dataset, we delimit points by '.' and ':'"""

def point_to_list(point):
    return re.split('\.|:', point)
    
point_paths = [point_to_list(p) for p in points]
pprint.pprint(point_paths)
```

```python
"""let's put those in a tree, treating each delimited point as the path to a leaf."""

point_tree = PointTree()
for point in points:
    point_path = point_to_list(point)
    point_tree.add_to_tree(point_path)

point_tree
```

```python
"""
Note that if we print the paths for the leaf nodes, they look almost identical to the paths.
The only diff is the points contain some none leaf nodes.
"""
nodes = point_tree.get_all_nodes()
leaf_nodes = [n for n in nodes if n.children ==[]]
leaf_node_paths = [n.get_path() for n in leaf_nodes]

concat_leaf_node_paths = [".".join(p) for p in leaf_node_paths]
concat_point_paths = [".".join(p) for p in point_paths]

print(set(concat_point_paths).difference(set(concat_leaf_node_paths)))
print(set(concat_leaf_node_paths).difference(set(concat_point_paths)))
```

```python
"""
Now that we have all the points represented in the tree, we want to asign brick classes to as many of
them as we can. to get the ball rolling, here's the unique set of node names we'd have to map.
"""

unique_node_names = set([n.name for n in nodes])

len(unique_node_names), unique_node_names
```

```python
"""
That's a lot, but we got a few tricks we can pull here...
a regex white list/black list of "brickable" names,
a regex white list/black list of rooms,
removing the number defining unique AHUs, we have options from here. 
"""

unique_node_names_no_numbers = set([re.sub("\d+", "", n) for n in unique_node_names])


# for n  in unique_node_names:
#     print(n, re.sub("\d+", "", n))

len(unique_node_names_no_numbers), unique_node_names_no_numbers
```
